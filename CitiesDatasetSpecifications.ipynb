# Census Data
Census data is located at https://github.com/kschnippel/Census/tree/master/2018_National_and_State_Population_Estimates/Census_Measure

I used 2018 because it is most recent dataset with place FIPS CODE

```python
# Preselect column type as object
coltype = {'SUMLEV':object,
          'STATE':object,
           'COUNTY':object,
          'PLACE':object,
          'COUSUB':object,
          'CONCIT':object,
          'PRIMGEO_FLAG':object}
```

```python
# import Census Data

census_2018 = pd.read_csv("../2018_National_and_State_Population_Estimates/Census_Measure/sub-est2018_all.csv",\
                         encoding="ISO-8859-1", dtype= coltype)
```

```python
census_2018.head()
```

## To do list
1. filter out place only
2. Creating 7 digit FIPS CODE column, take out other un-used columns.
3. Year
4. check the city is belonged to 500 cities in City Health dataset

```python
# To find place FIPS code, SUMLEV is 162 for place code.

city_df = census_2018[census_2018['SUMLEV']=='162']
```

```python
# 2. Creating 7 digit FIPS CODE column, take out other un-used columns.
# get column name first
city_df.columns
```

Removing columns
COUNTY,COUSUB', 'CONCIT','PRIMGEO_FLAG', 'FUNCSTAT','CENSUS2010POP','ESTIMATESBASE2010'
For popestimate, the following are sources of the data.

popestimates are resident pop. estimate in 7/1 of each year.
https://github.com/kschnippel/Census/blob/master/2018_National_and_State_Population_Estimates/Census_Measure/sub-est2018.pdf


```python
# Only necessary column
city_df = city_df[['SUMLEV', 'STATE', 'PLACE', 
       'NAME', 'STNAME', 'POPESTIMATE2010', 'POPESTIMATE2011',
       'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014',
       'POPESTIMATE2015', 'POPESTIMATE2016', 'POPESTIMATE2017',
       'POPESTIMATE2018']]
```

```python
city_df.head()
```

```python
# creating CITY_FIPS_CODE columm

city_df['CITY_FIPS_CODE'] = city_df['STATE']+city_df['PLACE']
```

```python
city_df.head()
```

```python
city_df.columns
```

```python
# 3 each year

city_df = city_df.melt(id_vars=['SUMLEV', 'STATE', 'PLACE', 'NAME', 'STNAME','CITY_FIPS_CODE'], 
        var_name="Year", 
        value_name="Pop_Estimate")
```

```python
# last 4 word from year is year

city_df['Year']=city_df['Year'].apply(lambda x: str(x)[-4:])
```

```python
city_df
```
